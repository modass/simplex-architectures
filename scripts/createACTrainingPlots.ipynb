```python
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300

x = [x+1 for x in range(10)]
ac_simple_trainings = [67,16,3,9,5,2,3,1,1,1]
ac_adv_trainings = [345,52,11,12,7,8,4,6,7,1]

ac_simple_bc = [12,9,9,6,9,7,7,7,8,8]
ac_adv_bc = [55,48,46,46,43,42,43,41,40,41]

fix, ax1 = plt.subplots()
ax2 = ax1.twinx()

l1 = ax1.plot(x,ac_simple_trainings, label='AC1 trainings', color='blue', linestyle='--')
l2 = ax1.plot(x,ac_adv_trainings, label='AC2 trainings', color='orange', linestyle='-.')
l3 = ax2.plot(x,ac_simple_bc, label='AC1 BC invok.', color='blue', alpha=0.3, linestyle=':')
l4 = ax2.plot(x,ac_adv_bc, label='AC2 BC invok.', color='orange', alpha=0.3)

ax1.set_xlabel('laps')
ax1.set_ylabel('#trainings')
ax2.set_ylabel('#BC invocations')

lns = l1+l2+l3+l4
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=0)

plt.savefig('trainings_over_time.png')
plt.show()
```
